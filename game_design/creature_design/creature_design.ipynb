## Player Status

```python
from graphviz import Digraph

dot = Digraph('player status', filename='player_status')
dot.attr(label='Player Status', splines='polyline', rankdir='LR')

dot.attr('node', shape='oval')
dot.node('player', 'Player')


with dot.subgraph() as s:
    s.attr('node', shape='box', rank='same')
    s.node('attr', 'Attributes')
    s.node('name', 'Name')
    s.node('hp', 'Health Value')
    s.node('armor', 'Armor')
    

with dot.subgraph() as s:
    s.attr('node', shape='box', rank='same')
    s.node('stre', 'Strength')
    s.node('inte', 'Intelligence')
    s.node('agi', 'Agility')
    s.node('defe', 'Defence')
    s.node('fai', 'Faith')
    s.node('san', 'Sanity')
    s.node('luc', 'Luck')
    s.node('head', 'Head-wear')
    s.node('body', 'Body-wear')
    s.node('weap', 'Weapon')
    s.node('foot', 'Foot-wear')

dot.edge('player', 'attr')
dot.edge('player', 'name')
dot.edge('player', 'hp')
dot.edge('player', 'armor')
dot.edge('attr', 'stre')
dot.edge('attr', 'inte')
dot.edge('attr', 'agi')
dot.edge('attr', 'defe')
dot.edge('attr', 'fai')
dot.edge('attr', 'san')
dot.edge('attr', 'luc')
dot.edge('armor', 'head')
dot.edge('armor', 'body')
dot.edge('armor', 'weap')
dot.edge('armor', 'foot')

# preview
dot
```
